### One-Hot Encoding of categrical variables( job,marital,education,contact,poutcome)


```python
#We're using ONE-HOT Encoder to add value count of categorical variables into columns.
```

```python
Bank_One_Hot=pd.get_dummies(data=Bank_data_1,columns=['job','marital','education','contact','poutcome'])
Bank_One_Hot
```

```python
#To see all columns of value count
pd.set_option("display.max_columns",None)
Bank_One_Hot
```

```python
Bank_One_Hot.info()
```

```python
#Still,5 objects are categorical we'he to convert into numerical values.
# Custom Binary Encoding of Binary o/p variables(for Yes/No type value counts) like Default,Housing,loan & y. '4=Objects'
```

```python
Bank_One_Hot['default']=np.where(Bank_One_Hot['default'].str.contains("Yes"),1,0)
Bank_One_Hot['housing']=np.where(Bank_One_Hot['housing'].str.contains("Yes"),1,0)
Bank_One_Hot['loan']=np.where(Bank_One_Hot['loan'].str.contains("Yes"),1,0)
Bank_One_Hot['y']=np.where(Bank_One_Hot['y'].str.contains("Yes"),1,0)
Bank_One_Hot
```

```python
Bank_One_Hot.info()
```

```python
#Now,all variables in Numerical Form except 'Month' Variable.
# Find and Replace Encoding for month categorical varaible
Bank_One_Hot['month'].value_counts()
```

```python
Data={'month':{'jan': 1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov': 11,'dec':12}}
```

```python
Bank_One_Hot=Bank_One_Hot.replace(Data)
Bank_One_Hot
```
